# OperationContainer Traceback Demo
This notebook demonstrates how to use OperationContainer to display tracebacks in a user-friendly way.

```python
import os
import sys
import importlib.util
# Path to the demo file
demo_path = os.path.join(os.getcwd(), 'demos/operation_container_traceback_demo.py')

# Load the module directly from file
spec = importlib.util.spec_from_file_location("demo_module", demo_path)
demo_module = importlib.util.module_from_spec(spec)
sys.modules["demo_module"] = demo_module
spec.loader.exec_module(demo_module)

# Get the run_demo function
run_demo = demo_module.run_demo
```

```python
# Run the demo
demo_ui = run_demo()
display(demo_ui)
```

## How to Use
1. Click the buttons above to generate different types of errors
2. Observe how the tracebacks are displayed in the log accordion
3. Use the 'Clear Logs' button to clear the log messages
